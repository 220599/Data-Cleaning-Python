```python
import pandas as pd
```

```python
df = pd.read_excel('/content/Customer Call List.xlsx')
```

```python
df
```

```python
df.info()
```

```python
df = df.drop_duplicates()
```

```python
df = df.drop(columns = 'Not_Useful_Column')
```

```python
df
```

```python
#df["Last_Name"] = df["Last_Name"].str.lstrip("/")
#f["Last_Name"] = df["Last_Name"].str.lstrip("...")
#df["Last_Name"] = df["Last_Name"].str.rstrip("_")
#df["Last_Name"]
df["Last_Name"] = df["Last_Name"].str.strip("123./_")
df["Last_Name"]
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','',regex=True)
df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df["Phone_Number"]= df["Phone_Number"].apply(lambda x: str(x))
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df["Phone_Number"] = df["Phone_Number"].str.replace("nan--",'',regex=True)
df["Phone_Number"] = df["Phone_Number"].str.replace("Na--",'',regex=True)
df["Phone_Number"] = df["Phone_Number"].str.replace("--",'',regex=True)
df["Phone_Number"]
```

```python
df[["Street Address", "State", "Zip_Code"]] = df["Address"].str.split(',',n=2, expand = True)
df = df.drop(columns = "Address")
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N', regex=True)
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y', regex=True)
df['Do_Not_Contact'] = df['Paying Customer'].str.replace('No','N', regex=True)
df['Do_Not_Contact'] = df['Paying Customer'].str.replace('Yes','Y', regex=True)
df
```
